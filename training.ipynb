```python
import cv2
import csv
import os
import pandas as pd
from feat import Detector
from IPython.display import Image

from feat.utils import FEAT_EMOTION_COLUMNS
```

```python
import logging
logger = logging.getLogger()
logger.setLevel(logging.CRITICAL)
```

```python
detector = Detector(device="cuda", batch_size=20)
```

```python
df = pd.DataFrame()
for emotion in os.listdir("./train"):
    print(emotion)
    test_images = ["./train/" + emotion + "/" + file for file in os.listdir("./train/" + emotion)]
    test_images = test_images[400:800]
    prediction = detector.detect_image(test_images, batch_size=20)
    prediction.insert(0, "emotion", emotion)
    df = pd.concat([df, prediction])
    
```
